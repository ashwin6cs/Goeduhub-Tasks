### 8) Dealing with missing values

```python
df3 = pd.read_csv("/content/IMDB-Movie-Data.csv")
_ = msno.matrix(df3)
```

```python
mean_value1=df3['Revenue (Millions)'].mean()
mean_value2=df3['Metascore'].mean()
df3['Revenue (Millions)'].fillna(value=mean_value1, inplace=True)
df3['Metascore'].fillna(value=mean_value2, inplace=True)
df3.isnull().sum()
```

### 10) Apply( ) functions

```python
import pandas as pd

s = pd.read_csv("/content/IMDB-Movie-Data.csv")

def fun(num):

	if num<5:
		return "Low"

	elif num>= 5 and num<8:
		return "Normal"

	else:
		return "High"

new = s["Rating"].apply(fun)
del s["Rating"]
frames = [s,new]
pd.concat(frames,axis=1)
```
